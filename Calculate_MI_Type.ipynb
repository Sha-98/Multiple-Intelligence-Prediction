```python
def Classification(Data):
  a1 = ((Data['Logical_Agreement_sum'])*Data['Logical_Agreement_sum']/Data['Logical Score'])- ((Data['Logical_Disagreement_sum'])*Data['Logical_Disagreement_sum']/Data['Logical Score'])
  logical = np.array(a1)
  Data['Logical'] = logical

  a2 = ((Data['Musical_Agreement_sum'])*Data['Musical_Agreement_sum']/Data['Musical Score'])- ((Data['Musical_Disagreement_sum'])*Data['Musical_Disagreement_sum']/Data['Musical Score'])
  musical = np.array(a2)
  Data['Musical'] = musical

  a3 = ((Data['Naturalistic_Agreement_sum'])*Data['Naturalistic_Agreement_sum']/Data['Naturalistic Score'])- ((Data['Naturalistic_Disagreement_sum'])*Data['Naturalistic_Disagreement_sum']/Data['Naturalistic Score'])
  naturalistic = np.array(a3)
  Data['Naturalistic'] = naturalistic

  a4 = ((Data['Intrapersonal_Agreement_sum'])*Data['Intrapersonal_Agreement_sum']/Data['Intrapersonal Score'])- ((Data['Intrapersonal_Disagreement_sum'])*Data['Intrapersonal_Disagreement_sum']/Data['Intrapersonal Score'])
  intrapersonal = np.array(a4)
  Data['Intrapersonal'] = intrapersonal

  a5 = ((Data['Spatial_Agreement_sum'])*Data['Spatial_Agreement_sum']/Data['Spatial Score'])- ((Data['Spatial_Disagreement_sum'])*Data['Spatial_Disagreement_sum']/Data['Spatial Score'])
  spatial = np.array(a5)
  Data['Spatial'] = spatial

  a6 = ((Data['Linguistic_Agreement_sum'])*Data['Linguistic_Agreement_sum']/Data['Linguistic Score'])- ((Data['Linguistic_Disagreement_sum'])*Data['Linguistic_Disagreement_sum']/Data['Linguistic Score'])
  linguistic = np.array(a6)
  Data['Linguistic'] = linguistic

  a7 = ((Data['Interpersonal_Agreement_sum'])*Data['Interpersonal_Agreement_sum']/Data['Interpersonal Score'])- ((Data['Interpersonal_Disagreement_sum'])*Data['Interpersonal_Disagreement_sum']/Data['Interpersonal Score'])
  interpersonal = np.array(a7)
  Data['Interpersonal'] = interpersonal

  a8 = ((Data['Kinesthetic_Agreement_sum'])*Data['Kinesthetic_Agreement_sum']/Data['Kinesthetic Score'])- ((Data['Kinesthetic_Disagreement_sum'])*Data['Kinesthetic_Disagreement_sum']/Data['Kinesthetic Score'])
  kinesthetic = np.array(a8)
  Data['Kinesthetic'] = kinesthetic

  df02 = Data[['Logical', 'Musical', 'Naturalistic', 'Intrapersonal', 'Spatial','Linguistic', 'Interpersonal', 'Kinesthetic']]
  Intelligence = df02.idxmax(axis=1)
  Data['Intelligence_Predicted'] = Intelligence
```
